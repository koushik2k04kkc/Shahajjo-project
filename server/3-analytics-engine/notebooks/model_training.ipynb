# Z-Score and CV Threshold Tuning

This notebook is used to tune the parameters for the mathematical models in the analytics engine using mock historical transaction data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate mock transaction amounts
np.random.seed(42)
normal_amounts = np.random.normal(loc=500, scale=100, size=100)
anomaly_amounts = np.array([5000, 5200, 4800])
all_amounts = np.concatenate([normal_amounts, anomaly_amounts])

df = pd.DataFrame({'amount': all_amounts})
df['z_score'] = (df['amount'] - df['amount'].mean()) / df['amount'].std()

print(df.tail())
```
